###Método de bisección

```python
import matplotlib.pyplot as plt
import math as mt
import numpy as np
def f1(x):
    return np.exp(x) + 2**(-x)+2* mt.cos(x)-6
xa, xb , Eb , i, Errorb = 1, 2, 100, 0, [100]
xr = (xb+xa)/2
print("#iteración \t\t   xa \t\t  xb \t\t  xr \t\t  Error")
print("{0}\t\t\t {1:6.4f} \t {2:6.4f} \t {3:6.4f} \t {4:6.4f}".format(i, xa, xb, xr, Eb))
while Eb > .0001 :
  fa = f1(xa)
  fr = f1(xr)
  if fa*fr > 0:
    xA = xa
    xa = xr
    xr = (xb+xa)/2
    Eb = abs((xr-xa)/xr)
    Errorb.append(round(Eb*100,3))  
    print("{0}\t\t\t {1:5.4f} \t {2:5.4f} \t {3:5.4f} \t {4:5.4f}".format(i+1, xA, xb, xr, Eb))
  elif fa*fr < 0:
    xB = xb
    xb = xr
    xr = (xb+xa)/2
    Eb = abs((xr-xb)/xr)
    Errorb.append(round(Eb*100,3))
    print("{0}\t\t\t {1:5.4f} \t {2:5.4f} \t {3:5.4f} \t {4:5.4f}".format(i+1, xa, xB, xr, Eb))
  i += 1

print(f'La raíz en el intervalo [1,2] es {xr}')
```

###Método de Newton-Raphson

```python
def ff(x):
  return np.exp(x) - 2**(-x) * mt.log(2,mt.e) - 2*mt.sin(x)
if ff(xa) != 0 and ff(xb) !=0:
  En, j, xan, xbn, Errorn = 100, 0, 1, 2, [100]
  p = (xan + xbn)/2
  print("# Iteración \t P \t\t\t %Error")
  print("{0} \t\t {1:6.4f} \t\t {2:6.4f}".format(j, p , En)) 
  while En > .0001:
    p1 = p - (f1(p)/ff(p))
    En = abs(p1 -p)
    Errorn.append(round(En*100,3))
    p = p1
    j += 1
    print("{0} \t\t {1:6.4f} \t\t {2:6.4f}".format(j, p , En)) 

print(f'La raíz en el intervalo [{xan} , {xbn}] es {p}')
```

###Método de la secante

```python
Es, k, xas, xbs, Errors = 100, 0, 1, 2, [100]
p0 = (xas + xbs)/2
p1 = 2
print("# Iteración \t Pn \t\t\t %Error")
print("{0} \t\t {1:6.4f} \t\t {2:6.4f}".format(k, p0 ,  Es)) 
while Es > .0001:
  p2 = p1 - (f1(p1)*(p1-p0))/(f1(p1)-f1(p0))
  Es = abs(p2 -p1)
  Errors.append(round(Es*100,3))
  p0 = p1
  p1 = p2
  k += 1
  print("{0} \t\t {1:6.4f} \t\t {2:6.4f}".format(k, p2 , Es)) 
print(f'La raíz en el intervalo [{xas} , {xbs}] es {p0}')
```

###Método de Horner

```python
def definir (g):
  polinomio = {}
  print("\n")
  for t in range(g + 1):
    cof = float(input(f'Ingresa el coeficiente de x^{g- t}: '))
    polinomio[f'x^{g - t}']= cof
  return polinomio
```

```python
def sintetica (p, x0,gr):
  Q = {f'x^{gr}' : list(p.values())[0]}
  for m in range(gr+1):
    if m == gr:
        prim = list(Q.values())[m] * x0
        Q["r"] = prim + list(p.values())[m+1] 
    else:
        prim = list(Q.values())[m] * x0
        Q[f'x^{(gr-1) - m }'] = prim + list(p.values())[m+1] 
  return Q
```

```python
#General
grado = int(input("Ingresa el grado del polinomio: "))
P = definir(grado)
print("\n")
x0 = float(input("Ingrese el valor de x0: "))
print("\n")
n, E = 0, 1000
print(" n \t\t\t  Pn \t\t\t  Error")
print("{0} \t\t\t {1:6.4f} \t\t {2:6.4f}".format(n , x0, E))
while E > .000001:
  Q = sintetica(P,x0,grado-1)
  Q1 = sintetica(Q,x0,grado-2)
  M, N = list(Q.values())[grado], list(Q1.values())[grado-1]
  pn = round(x0 - (M/N),6)
  E = abs(x0-pn)
  x0 = pn  
  n += 1
  print("{0} \t\t\t {1:6.4f} \t\t {2:6.4f}".format(n , pn, E))
```

###Método de Muller

```python
def evaluar(p,valor, gr):
  suma = 0
  for m in range(gr + 1):
    prim = list(p.values())[m] * (valor**(gr-m))
    suma += prim
  return suma
```

```python
import cmath 
import math
grado = int(input("Ingresa el grado del polinomio: "))
f = definir(grado)
print("\n")
x0 = float(input("Ingrese el valor de x0: "))
x1 = float(input("Ingrese el valor de x1: "))
x2 = float(input("Ingrese el valor de x2: "))
f0, f1 ,f2, E, n0, n1 ,n2 =  evaluar(f,x0,grado), evaluar(f,x1,grado), evaluar(f,x2,grado), 100, 0, 1, 2

def inicial(x0,f0,x1,f1,x2,f2):
  d = (x0-x2)*(x1-x2)*(x0-x1)
  a = ((f0-f2)*(x1-x2)-(x0-x2)*(f1-f2))/d
  b = ((f1-f2)*(x0-x2)**2-(f0-f2)*(x1-x2)**2)/d
  return a,b

def general(a,b,c,xa):
  detp = abs(b + math.sqrt(abs(b**2 - 4*a*c)))
  detn = abs(b - math.sqrt(abs(b**2 - 4*a*c)))
  if detp > detn:
    xn = xa - (2*c)/(b + cmath.sqrt(b**2 - 4*a*c))
  elif detp < detn:
    xn = xa - (2*c/b - cmath.sqrt(b**2 - 4*a*c))
  return xn

print(" n \t\t\t  xn \t\t\t\t   fn \t\t\t\t  Error")
print("{0} \t\t\t {1:6.4f} \t\t\t {2:6.4f} \t\t\t  {3:6.4f}".format(n0 , x0, f0, E))
print("{0} \t\t\t {1:6.4f} \t\t\t  {2:6.4f} \t\t\t  {3:6.4f}".format(n1 , x1, f1, E))
print("{0} \t\t\t {1:6.4f} \t\t\t  {2:6.4f} \t\t\t  {3:6.4f}".format(n2 , x2, f2, E))

while E > .000001:
  f0 = evaluar(f,x0,grado)
  f1 = evaluar(f,x1,grado)
  f2 = evaluar(f,x2,grado)
  [x,y] = inicial(x0,f0,x1,f1,x2,f2)
  pn = general(x,y,f2,x2)
  fn = evaluar(f,pn,grado)
  E = abs(pn-x2)
  x0 = x1
  x1 = x2
  x2 = pn
  n2 += 1
  print("{0} \t\t\t {1:6.4f} \t\t {2:6.4f} \t\t {3:6.4f}".format(n2 ,pn, fn, E))
```
