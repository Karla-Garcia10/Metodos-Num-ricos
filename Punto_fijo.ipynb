```python
#Ingreso la función g(x)
import math 
def P(x):   
    return pow(10/(4+x),1/2)
# Ingreso el punto inicial 
p0 = 1.5
evaluado  = P(p0)

#Guardar números inicales
tol = float(input("Ingrese el valor de la tolerancia: "))
nmax = float(input("Ingrese el número máximo de iteraciones: "))
error , i = 100, 0

print("\n")
print("# iteración\t\t p0 \t\t P(p0) \t\t Error ")
print("{0} \t\t\t {1:6.4f} \t {2:6.4f} \t {3:6.4f}".format(i, p0, evaluado, error))

while i < nmax and error > tol:
  evaluado = P(p0)
  error = abs(evaluado - p0)
  if error > tol:
    p0 = evaluado
    man = P(p0)
    i += 1
    print("{0} \t\t\t {1:6.6f} \t {2:6.6f} \t {3:6.6f}".format(i, p0, man, error))

print("\n")
print(f'La raíz de la función es aproxiamadamente {round(man,6)}')

```
